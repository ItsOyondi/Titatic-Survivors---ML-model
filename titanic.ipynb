```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
#from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
#from sklearn.metrics import accuracy_score
```

```python
titanic = pd.read_csv("titanic.csv", index_col='PassengerId') #PassenderID used as index column
t_titanic = titanic.drop(columns=['Name','Parch','Ticket','Cabin', 'SibSp','Embarked'])
# t_titanic.head()
t_titanic.dropna(inplace=True) #remove all columns with NAN values
```

```python

# Change gender into numbers, 1 - Male, 2 Female
t_titanic['Sex'] = np.where((t_titanic['Sex'] == 'male'), 1, 2)

t_titanic
```

```python
#create a model
#model = LogisticRegression()
model =  DecisionTreeClassifier()
lab = preprocessing.LabelEncoder()

raw = t_titanic.drop(columns=['Survived'])
target = t_titanic['Survived']

target_transformed = lab.fit_transform(target) #solved the issue of continous-multioutput

#raw_train, target_train, raw_test, target_test = train_test_split(raw, target_transformed, test_size=0.2)
model.fit(raw, target_transformed)

predictions = model.predict([[2,2,48.0,50]]) #test the accuracy of the prediction model

predictions_res = np.where((predictions == 0), 'Did not survive', 'Survived') #Make the model print message of survived or not

predictions_res
#conclusion - the model accurately predicts the results.
#Funny findings from the model prediction results: Fare was a major determinant for the survival rate followed by gender and Pclass.
#Men who paid less fare did not survive compared to women with the same age and who paid the same fare. 
```
